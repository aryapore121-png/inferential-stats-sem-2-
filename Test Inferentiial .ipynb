```python
import numpy as np
from scipy.stats import chisquare
#observed as(x)
#expected as(y)
x = np.array([8, 6, 10, 7, 8, 11, 9])
y = np.array([9, 8, 11, 8, 10, 7, 6])

chi_stat, p_value = chisquare(f_obs=x, f_exp=y)

print("Chi-Square Statistic:", chi_stat)
print("p-value:", p_value)

#at 5% LOS
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Observed data does not fit.")
else:
    print("Fail to Reject H0: Observed data fits.")
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
n = 6   # number of tosses
p = 0.5 # probability of heads
k = 2   # number of heads

# Q2: Probability of exactly 2 heads
prob_2_heads = stats.binom.pmf(k, n, p)
print("Probability of 2 heads in 6 tosses:", prob_2_heads)

# Fit and plot binomial distribution
x = range(n+1)  # possible outcomes: 0 to 6 heads
pmf = [stats.binom.pmf(i, n, p) for i in x]

plt.bar(x, pmf, color='skyblue')
plt.title("Binomial Distribution (n=6, p=0.5)")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.show()
```
